```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
```

```python
# Plot gamma distribution with different parameters
x = np.linspace(0, 20, 100)
plt.figure(figsize=(12, 8))
plt.plot(x, stats.gamma.pdf(x, a=1, scale=1), label='a=1, b=1')
plt.plot(x, stats.gamma.pdf(x, a=2, scale=1), label='a=2, b=1')
plt.plot(x, stats.gamma.pdf(x, a=3, scale=1), label='a=3, b=1')
plt.plot(x, stats.gamma.pdf(x, a=3, scale=2), label='a=4, b=1')
plt.plot(x, stats.gamma.pdf(x, a=3, scale=1.5), label='a=5, b=1')
```
